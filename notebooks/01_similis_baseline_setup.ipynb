# SIMILIS baseline: setup

Импорт библиотек, версии окружения, фиксация `seed` и создание рабочей структуры папок.

```python
import os
import sys
import random
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from tqdm.auto import tqdm

import sklearn
import torch
import torchvision

try:
    import cv2
except ImportError:
    cv2 = None
```

```python
print(f'Python: {sys.version.split()[0]}')
print(f'Torch: {torch.__version__}')
print(f'Torchvision: {torchvision.__version__}')
print(f'Scikit-learn: {sklearn.__version__}')
print(f'OpenCV: {cv2.__version__ if cv2 is not None else "not installed"}')
print(f'CUDA available: {torch.cuda.is_available()}')
print(f'CUDA version: {torch.version.cuda}')
if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name(0)}')
```

```python
SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

os.environ['PYTHONHASHSEED'] = str(SEED)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

try:
    torch.use_deterministic_algorithms(True)
    deterministic_algorithms = True
except Exception:
    deterministic_algorithms = False

print(f'SEED fixed: {SEED}')
print(f'Deterministic algorithms: {deterministic_algorithms}')
```

```python
PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
PROJECT_ROOT
```

```python
required_dirs = [
    'data/raw',
    'data/interim',
    'data/processed',
    'splits',
    'artifacts/checkpoints',
    'artifacts/preds',
    'artifacts/reports',
    'artifacts/figures',
]

for rel_path in required_dirs:
    (PROJECT_ROOT / rel_path).mkdir(parents=True, exist_ok=True)

print('Created directories:')
for rel_path in required_dirs:
    print('-', rel_path)
```
